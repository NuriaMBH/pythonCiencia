```python
def multiplicarDoble(menciones):
    doble = menciones * 2
    return doble
#AGREGAMODS UNA NUEVA COLUMNA AL DATAFRAME CON EL DOBLE DE MENCIONES
df['doble_mentions'] = df['mentions'].apply(multiplicarDoble)
df.head()
```

vamos a crear una nueva columna que sea la popularidad del influencer, que esta compuesta en dividir los seguidores entre cuantas personas estoy siguiendo
(followees/followers)
para poder enviar la fila del df en lugar del valor de una columna se utiliza axis como parametro dentro de apply 
axis = 1--> significa enviar la fila

```python
def popularidad(fila): 
    followers = fila['followers']
    followees = fila['followees']
    resultado = followees/ followers
    return resultado
df['popularidad']= df.apply(popularidad,axis=1)
df[df['popularidad']>= 15]
df.head()
```

Vamos a agrupar por el campo COUNTRY Y y vamos a mostrar la media de dicho grupo.
dentro de nuestra df tenemos textos, no puede darme la media de danicarvajal

```python
df.groupby(['country']).agg({
    "followers": "sum",
    "mentions": "mean",
    "retweets": "max"
})
```

quiero mostrar los datos que tengan mas de 5.000 seguidaores

```python
grupo= df.groupby(['country']).agg({
    "followers": "sum",
    "mentions": "mean",
    "retweets": "max"
})
grupo[grupo['followers'] > 5000]
```

por supuesto , al igual que podemos leer datos, podemos exportar datos .vamos a almacenar el grupo dentro de un csv

```python
grupo.to_csv('data/grupo.csv')
grupo.to_json('data/grupo.json')
#grupo.to_excel(data/grupo.slsx')
```
