Ejemplo sobre IRIS que son datos de un tipo de flor

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/iris.data")
df
```

Dibujamos las últimas filas

```python
df.tail()
```

Podemos establecer columnas para nuestro DataFrame

```python
df.columns = ["longitud_sepalo", "ancho_sepalo", "longitud_petalo", "ancho_petalo", "clase"]
df
```

```python
df.describe()
```

Podemos contar valores por una columna

```python
df['clase'].value_counts()
```

Ordenar datos ascendente y descendente

```python
#df.sort_values('ancho_sepalo')
df.sort_values('ancho_sepalo', ascending=False)
```

Podemos cambiar filas por columnas

```python
df.T
```

Filtrar por una columna

```python
df[ df['longitud_petalo'] > 5 ]
```

Filtrar por más de una condición:
Se utiliza & para AND
Se utiliza | para OR

```python
df[ (df['longitud_petalo'] > 5) & (df['ancho_petalo'] < 1.5) ]
```

Comprobar valores perdidos dentro del DataFrame

```python
df.isna().sum()
```

Vamos a poner valores nan a alguna columna/fila

```python
df['longitud_petalo'][:2] = np.nan
df.head()
```

Mostrar los valores nulos

```python
df.isna().sum()
```

Podemos quitar los valores nulos SI sabemos dónde están (fillna)

```python
df['longitud_petalo'] = df['longitud_petalo'].fillna(2.9)
df.head()
```
